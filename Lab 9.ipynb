```python
# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
#Lab5 - actv. 7/8

query = 'SELECT *, IF(weekday(rental_date)<6,"workday","weekend") as day_type, WEEKDAY(rental_date) as checkday FROM rental'
display(query)
Actv_2 = pd.read_sql_query(query, engine)
Actv_2.head()
```

```python
# 3. Get all films with ARMAGEDDON in the title.
#Lab5 - actv. 10
query = 'SELECT title FROM film WHERE title LIKE "%%ARMAGEDDON%%"'
#print(query)
actv_3 = pd.read_sql_query(query, engine)
actv_3.head()
```

```python
# 4. Get 10 the longest films.
#Lab5 - actv. 3
query = '''SELECT title, length AS 'max duration'
FROM sakila.film
ORDER BY length DESC
LIMIT 10;'''
#print(query)
actv_4 = pd.read_sql_query(query, engine)
actv_4.head(10)
```

```python
# 5. How many films include Behind the Scenes content?
#Lab5 - actv. 13
query = '''SELECT COUNT(DISTINCT(title))
FROM sakila.film
WHERE special_features LIKE "%%Behind the Scenes%%"'''
#print(query)
actv_5 = pd.read_sql_query(query, engine)
actv_5.head()
```

```python
# 6. Which kind of movies (rating) have a mean duration of more than two hours?
#Lab 6 -actv. 7
query = '''SELECT rating, ROUND(AVG(length), 2) AS avg_length
FROM sakila.film
GROUP BY rating
HAVING avg_length>120'''
#print(query)
actv_6 = pd.read_sql_query(query, engine)
actv_6.head()
```

```python
# 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
#LAb 6 - actv8
query = '''SELECT title, length, RANK() OVER(ORDER BY length DESC) 
FROM sakila.film;'''
#print(query)
actv_7 = pd.read_sql_query(query, engine)
actv_7.head(60)
```
